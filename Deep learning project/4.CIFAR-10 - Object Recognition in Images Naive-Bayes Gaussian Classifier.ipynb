## 1.Load CIFAR-10 Database

```python
import keras
from keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

## 2.Visualize the first 40 training images

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(25,10))
for i in range(40):
    image = fig.add_subplot(4,10,i+1,xticks=[],yticks=[])
    image.imshow(np.squeeze(x_train[i]))
```

## 3.Rescaling the images

```python
x_train = x_train.astype('float32')/255
x_test = x_test.astype('float32')/255
```

## 4.Labels

```python
for i in range(5):
    print(y_train[i])
```

## 5.Reshaping datasets

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train=x_train.reshape(50000,3*32*32)
x_test=x_test.reshape(10000,3*32*32)
y_train = y_train.flatten()
y_test=y_test.flatten()
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## 6.Define the model architecture

```python
x_train.shape[1:]
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
```

## 7.Evaluate model classification accuracy on test set

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred)*100)
```
